```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
l=10 #change value accordingly
```

```python
def k(x1,x2):
    return math.exp(-(x1-x2)**2/(l**2))
```

```python
X=np.linspace(0,4*math.pi,100)
```

```python
Y=np.random.normal(0,math.sqrt(0.05),100)
Y=Y+np.sin(X)
```

```python
K=np.zeros((100,100))
for i in range(0,100):
    for j in range(0,100):
        K[i][j]=k(X[i],X[j])
```

```python
f=np.random.multivariate_normal(np.zeros(100),K,1)
```

```python
postmean= np.matmul(np.matmul(np.transpose(K),np.linalg.inv(K+0.05*np.identity(100))),Y)
```

```python
figure(figsize=(8,6))
plt.title('Value of $\ell =$'+str(l))
plt.plot(X,np.transpose(f),label='Prior Sample')
plt.plot(X,postmean,label = 'GP Posterior Mean')
plt.plot(X,np.sin(X),label='True function')
plt.legend()
```
